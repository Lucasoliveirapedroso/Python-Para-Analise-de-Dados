```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = range(1, 7) # Cria uma lista de valores de x de 1 a 6 
plt.plot(x, [xi * 2 for xi in x], label='2x') # cria um gráfico de linha com x no eixo x e 2 * x no eixo y (linha 1)
plt.plot(x, [xi * 3.0 for xi in x], label='3x') # cria outro gráfico de linha com x no eixo x e 3.0 * x no eixo y (linha 2)
plt.plot(x, [xi / 3.5 for xi in x], label='x/3.5')# cria mais um gráfico de linha com x no eixo x e x / 3.5 no eixo y (linha 3)
```

```python
plt.plot(x, [xi * 2 for xi in x], label='2x', marker='o')
plt.plot(x, [xi * 3.0 for xi in x], label='3x', marker='s')
plt.plot(x, [xi / 3.5 for xi in x], label='x/3.5', marker='^')
# Adicione uma legenda ao gráfico para identificar as linhas
plt.legend()
# Exiba o gráfico na tela
plt.show()
```

```python
plt.plot(x, [xi * 2 for xi in x], label='Legenda 1')
plt.plot(x, [xi * 3.0 for xi in x], label='Legenda 2')
plt.plot(x, [xi / 3.5 for xi in x], label='Legenda 3')
plt.grid(True)
plt.axis(xmin=1, ymax=20)
plt.xlabel('EIXO X')
plt.ylabel('EIXO Y')
plt.title(u'GRÁFICOS DE LINHA')
plt.legend()
plt.show()
```

```python
y = np.arange(1, 3) # Crie um array de valores de y de 1 a 2
plt.plot(y, '--', label='Série 1')  # Série 1 com estilo de linha tracejado ('--')
plt.plot(y + 1, '-.', label='Série 2')  # Série 2 com estilo de linha traço-ponto ('-.')
plt.plot(y + 2, ':', label='Série 3')  # Série 3 com estilo de linha pontilhado (':')
```

```python
y = np.arange(1, 3, 0.3) # Cria um array de valores de y de 1 a 3 com um intervalo de 0.3
plt.plot(y, "cx--", label='Série 1')  # Série 1: ciano (c), estilo de linha tracejado (--), marcador x (x)
plt.plot(y + 1, "mo:", label='Série 2')  # Série 2: magenta (m), estilo de linha traço-ponto (:), marcador círculo (o)
plt.plot(y + 2, "kp-.", label='Série 3')  # Série 3: preto (k), estilo de linha ponto-tracejado (-.), marcador quadrado (p)
```

```python
y = np.arange(1, 3, 0.2)
# Série 1 com marcadores 'x'
# Série 2 com marcadores 'o'
# Série 3 com marcadores 'D' (diamantes)
# Série 4 com marcadores '^' (triângulos)
# Série 5 com marcadores 's' (quadrados)
plt.plot(y,'x', y+0.5, 'o', y+1, 'D', y+1.5, '^', y+2, 's');
plt.show()
```
