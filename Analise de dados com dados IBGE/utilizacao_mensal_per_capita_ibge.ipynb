Tabela 1.1.38 - Rendimento, médio e mediano, mensal domiciliar per capita das pessoas de 10 anos ou mais de idade, por utilização da Internet, no período de referência dos últimos três meses, segundo as Grandes Regiões, as Unidades da Federação e as Regiões Metropolitanas - 2015

Rendimento médio mensal domiciliar per capita das pessoas de 10 anos ou mais de idade (R$) de utilização da Internet, no período de referência dos últimos três meses.


```python
import pandas as pd
```

```python
df = pd.read_excel('ibge.xlsx', 'Tab 1.1.38') # 'le arquivo excel', 'segundo lê panilha expecífica'

colNm = [
    'Regiao',
    'rem1',
    'utilizaram_mensal_per_capita',
    'nao_utilizaram_mensal_per_capta',
    'rem2',
    'rem3',
    'rem4',
] #colunas descritas / rem para remover que não vou utilizar
```

```python
df.columns = colNm

df = df.drop([0, 1, 2, 3, 46, 47]) #remove as linhas com índices 0, 1, 2, 3, 46 e 47 do DataFrame

df = df.drop(columns=['rem1','rem2','rem3','rem4' ]) # remove as colunas com os nomes 'rem1', 'rem2', 'rem3' e 'rem4' do DF

df = df.set_index(df['Regiao']) # define coluna 'Regiao' como índice 

df = df.drop(columns=['Regiao'])

df['utilizaram_mensal_per_capita'].mean() # calcula a média dos valores na coluna  
df['nao_utilizaram_mensal_per_capta'].mean()
df.plot.line(rot=90, figsize=(10,3)) # cria um gráfico de linha 
```

```python
# converte a coluna '[]' para um tipo numérico
df['utilizaram_mensal_per_capita'] = pd.to_numeric(df['utilizaram_mensal_per_capita']) 
df['nao_utilizaram_mensal_per_capta'] = pd.to_numeric(df['nao_utilizaram_mensal_per_capta'])

df.hist('utilizaram_mensal_per_capita') # gera um histograma dos valores na coluna
df.hist('nao_utilizaram_mensal_per_capta')
```
