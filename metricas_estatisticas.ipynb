```python
# importa bibliotecas
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# cria array de dados aleatórios 
dados = np.array([22, 25, 28, 30, 32, 15, 25, 14, 66, 35, 38, 40, 45, 50, 85, 38, 94, 10, 38, 38])
```

```python
# Média
media = np.mean(dados)
print(f"Média: {media:,.2f}")
```

```python
# Mediana
mediana = np.median(dados)
print(f"Mediana: {mediana}")
```

```python
# Moda
moda = stats.mode(dados, keepdims=True)
print(f"Moda: {moda.mode[0]} (aparece {moda.count[0]} vezes)")
```

```python
# Variância
variancia = np.var(dados)
print(f"Variância: {variancia:,.2f}")
```

```python
# Desvio Padrão
desvio_padrao = np.std(dados)
print(f"Desvio Padrão: {desvio_padrao}")
```

```python
# Intervalo
intervalo = np.ptp(dados)
print(f"Intervalo: {intervalo}")
```

```python
# Percentis
percentis = np.percentile(dados, [25, 50, 75])
print(f"25º Percentil: {percentis[0]}")
print(f"50º Percentil (Mediana): {percentis[1]}")
print(f"75º Percentil: {percentis[2]}")
```

```python
# Coeficiente de Correlação (exemplo com duas variáveis)
variavel1 = np.array([1, 2, 3, 4, 5, 34, 64, 73, 12, 43])
variavel2 = np.array([2, 4, 6, 8, 10, 49, 85, 84, 20, 69])
coeficiente_correlacao = np.corrcoef(variavel1, variavel2)[0, 1]
print(f"Coeficiente de Correlação: {coeficiente_correlacao}")

plt.figure(figsize=(8, 6)) #Cria uma nova figura com o tamanho especificado (8x6)
plt.scatter(variavel1, variavel2, color='blue', label='Dados') # Cria um gráfico de dispersão com variavel1 no eixo x, variavel2 no eixo y e pontos azuis
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')
plt.title('Gráfico de Dispersão')
plt.grid(True, linestyle='--', alpha=0.7) #  Adiciona uma grade ao gráfico

# Mostrando o gráfico
plt.legend()
plt.show()
```

```python
# Covariância (exemplo com duas variáveis)
covariancia = np.cov(variavel1, variavel2)[0, 1]
print(f"Covariância: {covariancia}")
```

```python
# Teste t de Student (exemplo de comparação de médias)
grupo1 = np.array([65, 70, 75, 80, 85])
grupo2 = np.array([55, 60, 65, 70, 75])
t_statistic, p_value = stats.ttest_ind(grupo1, grupo2)
print(f"Teste t de Student - Estatística t: {t_statistic}")
print(f"Valor-p: {p_value}")
```
