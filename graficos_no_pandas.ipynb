```python
import pandas as pd
import random as rd
```

```python
n_col = 3
n_row = 10
```

```python
df = pd.DataFrame()
```

```python
for i in range(1, n_col):
    col_rd = [rd.random() * 100 for i in range (0, n_row)]
    df["col"+str(i)] = col_rd
```

```python
df.hist('col1')
df.plot.line()
df.plot.area()
df.plot.area()
df.plot.pie(y = 'col1')
```

```python
df.plot.area()
df.plot.bar()
df.plot.barh()
df.plot.box()
df.plot.density()
df.plot.hexbin(x='col1', y='col2', gridsize=25)
df.plot.hist()
df.plot.kde()
df.plot.scatter(x='col1', y='col2')
```
