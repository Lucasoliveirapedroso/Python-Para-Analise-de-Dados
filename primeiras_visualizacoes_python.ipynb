```python
import random
import matplotlib.pyplot as plt
```

```python
x = range(1,50)
y = [random.random() * 100 for i in x]
```

```python
plt.plot(x,y) 
plt.grid(True)
plt.legend(['Dados aleat√≥rios'])
plt.title('Meu primeiro gr√°fico em Python')
plt.xlabel('S√©ries de dados')
plt.ylabel('Dados aleat√≥rios')
```

```python
plt.plot(x,y, color='black', linestyle='dashed') 
plt.grid(True)
plt.legend(['Dados aleat√≥rios'])
plt.title('Meu primeiro gr√°fico em Python')
plt.xlabel('S√©ries de dados')
plt.ylabel('Dados aleat√≥rios')
```
