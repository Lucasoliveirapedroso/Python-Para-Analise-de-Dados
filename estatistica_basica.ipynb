```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lê o arquivo Excel, "" tabela especfica
df = pd.read_excel("C:/Users/Lucas/OneDrive/Documentos/ibge.xlsx", 'Tab 1.1.38')

# Definir os nomes das colunas
colNm = [
    'Regiao',
    'MedioTotal',
    'MedioUtilizaram',
    'MedioNaoUtilizaram',
    'MedianoTotal',
    'MedianoUtilizaram',
    'MedianoNaoUtilizaram',
]
df.columns = colNm # Atribuir os nomes das colunas ao DataFrame
df = df.drop([0, 1, 2, 3, 46, 47]) # Remover as linhas com índices 0, 1, 2, 3, 46 e 47
df = df.set_index(df['Regiao']) # define coluna 'Regiao' como índice 
df = df.drop(columns=['Regiao']) # apaga a coluna regiao que não vai usuada que foi definida como indice
df = df.drop('Brasil') # Remover a linha com o rótulo 'Brasil'
df = df.sort_values('MedioTotal')# Ordenar o DataFrame por 'MedioTotal'
df.plot(rot=90) # Gera um gráfico de linhas com rotação dos rótulos em 90 graus
```

```python
# Converter todas as colunas para o tipo numérico, tratando erros como valores ausentes (NaN)
df = df.apply(pd.to_numeric, errors='coerce')

# Calcular a matriz de correlação entre as colunas do DataFrame
matCorr = df.corr()  # A matriz de correlação mostra como as variáveis se relacionam

# Gerar um gráfico de boxplot
df.boxplot(rot=90)  # rot=90 é usado para rotacionar os rótulos do eixo x em 90
```
